Câu hỏi 6: Liệt kê các trận đấu và thông tin về league và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
result_full_match = pd.read_sql(""" SELECT C.name AS 'Country',L.name AS 'League'
,M.season,M.stage,M.date
,M.match_api_id,M.home_team_api_id,M.away_team_api_id
FROM Match AS M
INNER JOIN Country AS C
ON M.country_id = C.id

INNER JOIN League AS L
ON M.league_id = L.id;""",conn)
result_full_match
```

7. Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
result_count_match = pd.read_sql(""" SELECT League.name, COUNT(*) as Count from Match, Country, League
WHERE Match.country_id = Country.id AND League.id = Match.league_id
GROUP BY League.name
ORDER BY Count DESC;""",conn)
result_count_match
```

12. Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp theo số lượng giảm dần

```python
result_name_goal = pd.read_sql("""SELECT team_short_name, Tongdiemhometeam+Tongdiemawayteam as Tongdiem FROM
(SELECT Team.team_short_name, SUM(Match.home_team_goal) as Tongdiemhometeam,SUM(Match.away_team_goal) as Tongdiemawayteam from Match,Team
WHERE Match.home_team_api_id = Team.team_api_id
GROUP BY Team.team_short_name)
ORDER BY Tongdiem DESC;""",conn)
result_name_goal
```
