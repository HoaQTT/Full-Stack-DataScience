##### Cân nhắc khi tải dữ liệu CSV

Có một số cân nhắc khi tải dữ liệu máy học của bạn từ tệp CSV.
Để tham khảo, bạn có thể tìm hiểu nhiều về kỳ vọng đối với tệp CSV bằng cách xem lại [CSV yêu cầu nhận xét có tiêu đề Định dạng chung và Loại MIME cho các giá trị được phân tách bằng dấu phẩy (CSV) Tệp] (https://tools.ietf.org/html/rfc4180)

**Tiêu đề tệp**

Dữ liệu của bạn có tiêu đề tệp không? Nếu vậy, điều này có thể giúp tự động gán tên cho từng
cột dữ liệu. Nếu không, bạn có thể cần đặt tên cho các thuộc tính của mình theo cách thủ công. Dù bằng cách nào, bạn nên
chỉ định rõ ràng liệu tệp CSV của bạn có tiêu đề tệp hay không khi tải dữ liệu của bạn.

**Bình luận**

Dữ liệu của bạn có nhận xét không? Nhận xét trong tệp CSV được biểu thị bằng dấu thăng (#) tại
đầu dòng. Nếu bạn có nhận xét trong tệp của mình, tùy thuộc vào phương pháp được sử dụng để tải
dữ liệu, bạn có thể cần cho biết có mong đợi nhận xét hay không và nhân vật sẽ mong đợi
để biểu thị một dòng bình luận.

**Dấu phân cách**

Dấu phân cách chuẩn phân tách các giá trị trong các trường là ký tự dấu phẩy (,). Tệp của bạn có thể
sử dụng một dấu phân tách khác như tab hoặc khoảng trắng trong trường hợp đó bạn phải chỉ định nó một cách rõ ràng.

**Báo giá**

Đôi khi giá trị trường có thể có khoảng trắng. Trong các tệp CSV này, các giá trị thường được trích dẫn. Các
ký tự trích dẫn mặc định là ký tự trong dấu ngoặc kép. Các ký tự khác có thể được sử dụng,
và bạn phải chỉ định ký tự trích dẫn được sử dụng trong tệp của mình.

# **Hiểu dữ liệu của bạn bằng hình ảnh hóa**

Bạn phải hiểu dữ liệu của mình để có được kết quả tốt nhất từ các thuật toán máy học. Cách nhanh nhất để tìm hiểu thêm về dữ liệu của bạn là sử dụng trực quan hóa dữ liệu. Trong phần này, bạn sẽ khám phá chính xác cách bạn có thể hình dung dữ liệu học máy của mình bằng Python bằng Pandas.

## Các lô đơn lẻ

Trong phần này, chúng ta sẽ xem xét ba kỹ thuật mà có thể sử dụng để hiểu từng thuộc tính của tập dữ liệu một cách độc lập.
1. Biểu đồ.

## Biểu đồ

Một cách nhanh chóng để có được ý tưởng về sự phân bố của từng thuộc tính là xem biểu đồ. Biểu đồ nhóm dữ liệu thành các thùng và cung cấp số lượng quan sát trong mỗi thùng. Từ hình dạng của các thùng, có thể nhanh chóng có được cảm giác liệu một thuộc tính có phải là Gaussian, bị lệch hay thậm chí có phân phối theo cấp số nhân hay không. Nó cũng có thể giúp nhìn thấy những ngoại lệ có thể xảy ra.

```python
import matplotlib.pyplot as plt

data.hist(figsize=(10,8))
plt.show()
```

Chúng ta có thể thấy rằng có lẽ các thuộc tính age, pedi và test có thể có phân phối theo cấp số nhân hoặc lệch. Chúng ta cũng có thể thấy rằng có lẽ các thuộc tính mass và pres và plas có thể có phân phối Gaussian hoặc gần như Gaussian. Điều này rất thú vị vì nhiều kỹ thuật học máy giả định một phân phối đơn biến Gaussian trên các biến đầu vào.

## Ô Mật độ
Biểu đồ mật độ là một cách khác để hiểu nhanh về sự phân bố của từng thuộc tính. Các ô trông giống như một biểu đồ trừu tượng với một đường cong mượt mà được vẽ qua đầu mỗi thùng, giống như mắt bạn đã cố gắng làm với các biểu đồ.

```python
data.plot(kind='density',subplots=True,layout=(3,3),sharex=False,figsize=(15,10))
plt.show()
```

Chúng ta có thể thấy sự phân bố cho từng thuộc tính rõ ràng hơn so với các biểu đồ.

## Ô và Ô Whisker
Một cách hữu ích khác để xem xét sự phân bố của từng thuộc tính là sử dụng Ô và Ô Whisker hay gọi tắt là ô hộp. Boxplots tóm tắt sự phân bố của từng thuộc tính, vẽ một đường cho giá trị trung bình (giá trị giữa) và một hộp xung quanh phân vị thứ 25 và 75 (50% ở giữa của dữ liệu). Râu đưa ra ý tưởng về sự lan truyền của dữ liệu và các dấu chấm bên ngoài râu hiển thị các giá trị ngoại lệ của ứng viên (giá trị lớn hơn 1,5 lần so với kích thước lan truyền của 50% dữ liệu ở giữa).

```python
data.plot(kind='box',subplots=True,layout=(3,3),sharex=False,sharey=False,figsize=(15,10))
plt.show()
```

Chúng ta có thể thấy rằng sự lây lan của các thuộc tính là khá khác nhau. Một số như tuổi tác, xét nghiệm và làn da
xuất hiện khá lệch về các giá trị nhỏ hơn.

## Lô đa biến
Phần này cung cấp các ví dụ về hai biểu đồ thể hiện sự tương tác giữa nhiều biến trong tập dữ liệu của bạn.
1. Đồ thị ma trận tương quan.
2. Ma trận đồ thị phân tán.

#### Lô ma trận tương quan
Tương quan cho biết mức độ liên quan của những thay đổi giữa hai biến. Nếu hai biến thay đổi theo cùng một hướng, chúng có tương quan thuận. Nếu chúng thay đổi theo các hướng ngược nhau với nhau (một đi lên, một đi xuống), thì chúng có tương quan nghịch. có thể tính toán mối tương quan giữa từng cặp thuộc tính. Đây được gọi là ma trận tương quan. Sau đó, có thể vẽ biểu đồ của ma trận tương quan và có được ý tưởng về những biến nào có mối tương quan cao với nhau. Điều này rất hữu ích cần biết vì một số thuật toán học máy như hồi quy tuyến tính và logistic có thể có hiệu suất kém nếu có các biến đầu vào tương quan cao trong dữ liệu.

```python
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
```

#### Ma trận Lô phân tán

Việc vẽ tất cả các biểu đồ phân tán này lại với nhau được gọi là ma trận biểu đồ phân tán. Biểu đồ phân tán hữu ích để phát hiện các mối quan hệ có cấu trúc giữa các biến, chẳng hạn như liệu có thể tóm tắt mối quan hệ giữa hai biến bằng một đường thẳng hay không. Các thuộc tính có mối quan hệ có cấu trúc cũng có thể tương quan và là ứng cử viên tốt để xóa khỏi tập dữ liệu.
